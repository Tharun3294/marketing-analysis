```python
import pandas as pd
```

```python
df["Income"]= df["Income"].fillna(df["Income"].median())
```

```python
df.head()
```

```python
# Create Age from Year_Birth
df["Age"] = 2025-df["Year_Birth"]
```

```python
df.head()
```

```python
# Total number of kids/teens at home
df["Kids_Teen_Home"]=df["Kidhome"]+df["Teenhome"]
```

```python
df.head()
```

```python
# Total Amount Spent
df["TotalSpend"]=(df["MntWines"]+df["MntFruits"]+df["MntMeatProducts"]+df["MntFishProducts"]+df["MntSweetProducts"]+df["MntGoldProds"])
```

```python
df.head()
```

```python
def categorize_age(age):
    if age < 30:
        return "Young"
    elif 30 <= age < 50:
        return "Middle-Aged"
    else:
        return "Senior"

df["Age_Group"]=df["Age"].apply(categorize_age)
```

```python
df.head()
```

```python
from datetime import datetime
# Convert 'Dt_Customer' to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
# Create "Seniority" in months
today = datetime(2025, 4, 27)
df['Customer_Seniority_Months'] = ((today - df['Dt_Customer']).dt.days) // 30
```

```python
df.head()
```

```python
# Create Spending Tier
def spending_tier(spend):
    if spend < 500:
        return "Low"
    elif 500 <= spend < 2000:
        return "Medium"
    else:
        return "High"

df["Spending_Tier"] = df["TotalSpend"].apply(spending_tier)
```

```python
df.head()
```

```python
campaigns = ["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5"]
df["Total_Accepted_Campaigns"] = df[campaigns].sum(axis=1)
```

```python
df['Web_Engagement_Score'] = df['NumWebPurchases'] * 2 - df['NumWebVisitsMonth']
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Income'], bins=30, kde=True, color='salmon')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Response', y='Age', data=df)
plt.title('Response Rate by Age')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Marital_Status', hue='Response', data=df)
plt.title('Campaign Response by Marital Status')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Response', y='TotalSpend', data=df)
plt.title('Spending Behavior by Campaign Response')
plt.show()
```

```python
# Campaign Acceptance Rate
acceptance_rate = df['Response'].mean() * 100
print(f"Overall Campaign Acceptance Rate: {acceptance_rate:.2f}%")

# Average Spending
average_spending = df['TotalSpend'].mean()
print(f"Average Total Spending per Customer: {average_spending:.2f}")

# Average Recency
average_recency = df['Recency'].mean()
print(f"Average Recency (days since last purchase): {average_recency:.2f}")
```

```python
#Customer Seniority vs Response (loyalty check)
plt.figure(figsize=(8,5))
sns.boxplot(x='Response', y='Customer_Seniority_Months', data=df, palette='pastel')
plt.title('Customer Seniority by Campaign Response')
plt.xlabel('Accepted Campaign')
plt.ylabel('Seniority in Months')
plt.show()
```
